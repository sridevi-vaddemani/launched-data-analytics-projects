```python
from scipy.stats import norm
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.hist(data1.Height,bins=20,rwidth=0.8,density=True)
plt.xlabel('Height (inches)')
plt.ylabel('Count')

rng = np.arange(data1.Height.min(),data1.Height.max(),0.1)
plt.plot(rng,norm.pdf(rng,data1.Height.mean(),data1.Height.std()))
```

```python
data1.Height.mean()
```

```python
upper_limit = data1.Height.mean() + 3 * data1.Height.std()
upper_limit
```

```python
lower_limit = data1.Height.mean() - 3 * data1.Height.std()
lower_limit
```

```python
 data1[(data1.Height>upper_limit )| (data1.Height <lower_limit)]
```

```python
data1_no_outlier_std_dev = data1[(data1.Height<upper_limit ) & (data1.Height >lower_limit)]
data1_no_outlier_std_dev.shape
```

```python
data1.shape[0]- data1_no_outlier_std_dev.shape[0]
```

```python
data1['zscore'] = (data1.Height- data1.Height.mean())/data1.Height.std()
```

```python
data1.head(5)
```

```python
data1[(data1['zscore']>3) | (data1['zscore']<-3)]
```

```python
data1_no_outliers = data1[(data1.zscore <3) & (data1.zscore >-3)]
data1_no_outliers.head
```

```python
data1.shape[0]- data1_no_outliers.shape[0]
```
